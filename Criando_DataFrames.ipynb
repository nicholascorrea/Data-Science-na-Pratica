# Criando DataFrames

## DataFrames a partir de dicionários
Uma das estruturas básicas do Python é o Dicionário. É muito conveniente informarmos nossos dados em uma variável do tipo `dict`, pois sua conversão em *DataFrame* é muito simples e direta.

```python
# importando o pandas
import pandas as pd
```

```python
# criar dicionário
dados = {
    'nome': ['Bart', 'Homer', 'Lisa', 'Marge'],
    'idade': [10, 45, 15, 40],
    'cidade': ['São Paulo', 'Salvador', 'Rio de Janeiro', 'Curitiba'],
    'comprou': [True, False, False, True]
}
```

```python
#criar DataFrame
df = pd.DataFrame(dados)
```

```python
#visualizar DataFrame
df
```

Após criar o DataFrame, podemos adicionar elementos a ele.
Por exemplo, adicionar uma lista de id de usuários ao DataFrame como no exemplo abaixo.

```python
# criar id do usuario
id_usuario = [12, 20, 55, 15]
```

```python
# associar id do usuário a lista
df.index = id_usuario
```

```python
# Visualizar o DataFrame
df
```

## DataFrames a partir de listas

para organizar as listas de modo que seja fácil converter para *DatFrame*, podemos usar a função *built-in* `zip()`

```python
# listas
dados = [['Carlos', 35, 'Araraquara', 'True'],
         ['Theo', 32, 'Natal', 'False'],
         ['Raquel', 12, 'SJC', 'True'],
         ['Fernanda', 14, 'Brasilia', 'True']]

# criar DataFrame
df = pd.DataFrame(dados, columns=['nome', 'idade', 'cidade', 'comprou'],
                  index=[124, 543, 999, 214])

# visualizar DataFrame
df
```

Podemos tilizar a função `loc` para selecionar linhas através do index

```python
df.loc[124]
```

## Criar novas colunas
Uma maneira simples de criar novas colunas em um *DataFrame* é usar a funcionalidade do Pandas conhecida como *broadcasting*

Informando o nome da nova coluna e declarando apenas um valor, ele será replicado para todas as linhas do *DataFrame*

```python
# criar nova coluna saldo
df['saldo'] = 0.0

# visualizar DataFrame
df
```

## Modificar *index* e colunas

Podemos alterar o nome dos índices ou o *label* das colunas

É possível alterar esse atributo da variável acessando diretamente:

```python
# alterar o index do df
df.index = ['a', 'b', 'c', 'd']

# visualizar DataFrame
df
```

```python
df.loc['a']
```

Para alterar o nome das colunas, posso alterar diretamente no atributo da variável:

```python
# alterar os labels das colunas
df.columns = ['Nome do Cliente', "Idade", "Naturalidade", "Fez Compras?", "Saldo na Loja"]

# visualizar DataFrame
df
```
